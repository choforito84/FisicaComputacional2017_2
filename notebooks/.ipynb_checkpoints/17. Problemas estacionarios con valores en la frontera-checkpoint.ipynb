Hasta ahora, hemos estudiado únicamente las **ecuaciones de evolución**, o sea, un problema **con valores iniciales**, en el cual se quiere determinar la evolución temporal de la solución, tanto para EDOs como para EDPs.

Otra clase fundamental de problemas son los **problemas estacionarios** o **problemas estáticos**, también llamados **problemas con valores en la frontera**. En este tipo de problemas, se busca una solución estacionaria, es decir, una solución que *no* cambia en el tiempo, pero que depende del espacio.

Este tipo de problemas surge en distintos contextos. Por ejemplo, si resolvemos la ecuación de difusión con condiciones de Dirichlet dadas que no sean cero, la solución temporal *converge* a una solución estacionaria. ¿Podemos calcular directamente esta solución estacionaria? Los problemas estacionarios también ocurren naturalmente al resolver, por ejemplo, las ecuaciones de Laplace y de Poisson para determinar el potencial (y de ahí el campo) eléctrico para una configuración dada de cargas y fronteras.

## Las ecuaciones de Poisson y de Laplace 

Consideremos el problema de encontrar el campo de potencial debido a una distribución de carga eléctrica. El campo de potencial $V(\mathbf{x})$ satisface la ecuación de Poisson,

$$ \nabla^2 V(\mathbf{x}) = -\rho(\mathbf{x}), $$

donde $\rho(\mathbf{x})$ es la densidad de carga en unidades adecuadas.

con los valores en la frontera  $V = V_1$ y $V = V_2$ en las dos placas.

Para comenzar, reduzcamos el problema a una sola dimensión. Entonces la ecuación se vuelve

$$ \frac{d^2 V(x)}{dx^2} = -\rho(x) \quad (*) $$ 

para todas las $x$,
con los potenciales en las placas $V(a) = x_1$ y $V(b) = x_2$,
es decir, una EDO con valores en la frontera.

Este tipo de problema es de una naturaleza *completamente diferente* de los que hemos visto anteriormente en EDOs, ya que *no hay evolución* en el tiempo. Y sin embargo...

Aproximando la segunda derivada con una diferencia centrada, se obtiene la siguiente discretización para la ecuación:

$$ \frac{V_{i+1}+V_{i-1}-2V_{i}}{h^2}=\rho_i $$

Esta ecuación se debe de cumplir de manera simultánea para todos los valores de $i$. Podemos reescribir este utilizando una matriz, con lo cual el problema adquiere la siguiente forma:

$$\mathbf{ \rho}=\mathbf{A} V $$

Donde A es una matriz tal que 

$$A_{i,i}=\frac{-2}{h^2}, \quad A_{i,i+1}=A_{i+1,i}=\frac{1}{h^2}$$

Hay varios métodos posibles para resolver problemas de esta índole. Empezaremos con ocupar un método **iterativo** que recién vimos:

```julia
function deltadirac(x;inf::Float64=1e2)
    if norm(x)<1/inf
        return inf
    else
        return 0.0
    end
end
```

```julia
A=linspace(-1,1,200)
plot(A,[deltadirac(a) for a in A])
```

```julia
function poisson(f::Function,L::Tuple{Float64,Float64},VL::Tuple{Float64,Float64};N::Int64=100,k::Float64=1e-2,test=false)
    X=collect(L[1]:k:L[2])
    V=copy(X)
    rho=[f(x) for x in X]
    for i in 1:N
        A=[V[i] for i in 1:(length(V)-1)]
        unshift!(A,VL[1])
        B=[V[i] for i in 2:(length(V))]
        push!(B,VL[2])
        if test==true
            @show A
            @show B
            @show V
            @show rho
            @show ((k^2))*rho
            @show 1/2*((k^2)*rho+A+B)
            println("\n\n")
        end
        V=1/2*((k^2)*rho+A+B)
    end
    return (X,V)
end
```

```julia
aldo=poisson(x->0,(-1.0,1.0),(1.0,1.0),N=50000,k=1e-2)
plot(aldo[1],aldo[2],ylim=(0.0,1.5),title="Potencial sin carga",xlabel="x",ylabel="V(x)")
```

```julia
aldo=poisson(x->0,(-1.0,1.0),(3.0,2.0),N=50000,k=1e-2)
plot(aldo[1],aldo[2],ylim=(0.0,4),title="Potencial sin carga",xlabel="x",ylabel="V(x)")
```

```julia
function poisson2(L::Tuple{Float64,Float64},VL::Tuple{Float64,Float64};N::Int64=100,k::Float64=1e-2,test=false)
    X=collect(L[1]:k:L[2])
    V=zeros(X)
    V[1]=VL[1]
    V[end]=VL[2]
    for i in 1:N
        A=[V[i] for i in 1:(length(V)-1)]
        unshift!(A,VL[1])
        B=[V[i] for i in 2:(length(V))]
        push!(B,VL[2])
        if test==true
            @show A
            @show B
            @show V
            @show rho
            @show ((k^2))*V
            @show 1/2*((k^2)*V+A+B)
            println("\n\n")
        end
        V=1/2*((k^2)*V+A+B)
    end
    return (X,V)
end
```

```julia
sayeg=poisson2((0.0,pi/2),(0.0,2.0),N=100000)
plot(sayeg[1],sayeg[2],label="analítica",title="solución a la ecuación planteada")
plot!(sayeg[1],[2*sin(x) for x in sayeg[1]],linestyle=:dash,label="numérica")
```

```julia
sayeg=poisson2((0.0,pi/2),(1.0,2.0),N=200000)
plot(sayeg[1],sayeg[2],label="analítica",title="solución a la ecuación planteada",linewidth=5)
plot!(sayeg[1],[2*sin(x)+cos(x) for x in sayeg[1]],linestyle=:dash,label="numérica",linewidth=5)
```
