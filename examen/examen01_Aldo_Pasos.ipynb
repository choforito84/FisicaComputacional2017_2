(i) Utiliza un bucle `for` para escribir una función `log_factorial` que toma un argumento `n` y que calcula el logaritmo del factorial de `n`. Para hacerlo, recuerda que el logaritmo de un producto es la suma de los logaritmos. [La función "logaritmo" en Julia se llama `log`.]

(ii) Utiliza una comprensión de arreglo para calcular $\log(n!)$ para $n$ de 1 a 100.

(iii) Dibuja esta cantidad en función de $n$. ¿Cualitativamente, cuál parece ser su tasa de crecimiento? ¿Qué implica esto sobre la tasa de crecimiento de $n!$?

```julia
using Plots
gr()
```

```julia
#[1]
function log_factorial(n::Int64)
    if n<0
        error("No se puede calcular el factorial de un numero negativo")
    end
    total=log(1)
    for i in 2:n
        total+=log(i)
    end
    return total
end
A=[log_factorial(x) for x in 1:100]
scatter(1:100,A,title="log \_factorial",xlabel="n",ylabel="log(n!)")
```

```julia
println(log_factorial(50))
println(log_factorial(100))
```

[2] Considera la suma de los cuadrados de los primeros $n$ enteros positivos,

$$S_n := \sum_{i=1}^n i^2.$$

Escriba una función que acepta un número $M$ y encuentra el valor mínimo de $n$ tal que $S_n > M$, es decir, tal que la suma exceda el valor dado.

```julia
#[2]
function suma_enteros(M)
    i=1
    total=0
    while M>total
        total+= i^2
        i+=1
    end
    return total
end

suma_enteros(15)
```

```julia
#[3]
function potencias1(x,tol=1e-10)
    total=0
    i=1
    termino=x
    if abs(x)>=1
        error("la función no converge en esos valores")
    end
    while abs(termino)>tol
        total+= termino
        termino*= x*(i/(i+1))
        i+=1
    end
    return total
end
println(potencias(-.5))
println(-log(1.5))
```

```julia
A=-.9:.01:.9
plot(A,[potencias1(x) for x in A],title="potencias",xlabel="x",ylabel="f(x)",)
```

```julia
plot(A,[potencias1(x) for x in A],title="potencias",xlabel="x",ylabel="f(x)",label="potencias1",linewidth=8,alpha=.4)
plot!(A,[-log(1-x) for x in A],label="-log(1-x)",linewidth=8,alpha=.4,linestyle=:dash)
```

$$ e^{-f(x)} \approx e^{-(-log(1-x))} = e^log(1-x)} = 1-x $$

```julia
B=.01:.005:.99
plot(B,[e^(-potencias1(x)) for x in B],label="e^(-potencias1(x))",linewidth=6,alpha=.6)
plot!(B,[1-x for x in B],label="1-x",linewidth=6,alpha=.6,linestyle=:dash)
```
